```python
from sympy import *
from sympy.plotting import plot as plt
import matplotlib 
init_printing()
plt.set_linestyle='dashed'
```

```python
x = Symbol('x')
```

```python
y=-5*x**2+10*x+150
```

```python
plt(y,(x, -20, 30), show=True)
print(y)
```

```python
diff(y)
```

```python
plt(diff(y))
```

```python
x1, x2 = solve(y)
```

```python
type(solve(y))
```

```python
round(x1.evalf(),3)
```

```python
round(x2.evalf(),2)
```
